```python
import numpy as np 
import pandas as pd 
```

```python
crimes = pd.read_csv('Chicago_Crimes_2012_to_2017.csv',error_bad_lines=False)
```

```python
crimes.head()
```

```python
#crimes.drop(['Unnamed: 0', 'Case Number','Description','Location Description' ,'Block','IUCR', 'X Coordinate', 'Y Coordinate','Updated On','Year', 'FBI Code', 'Beat','Ward','Community Area', 'Location', 'District'], inplace=True, axis=1)
```

```python
crime_light=crimes[['Date','Primary Type','Arrest']]
```

```python
crime_light.head()
```

```python
crime_light['year'] = pd.DatetimeIndex(crime_light['Date']).year
```

```python
crime_light.drop(['Date'],inplace=True, axis=1)
```

```python
crime_light.to_csv('Chicago_Crimes_2012_to_2016_light.csv', index=False)
```

```python
crime_light['month'] = pd.DatetimeIndex(crime_light['Date']).month
```

```python
crime_light = crime_light[crime_light.year != 2017]
```

```python
crime_light.shape
```

```python
crime_light.drop(['year'],inplace=True, axis=1)
```

```python
crime_light.head()
```

```python
df3=crime_light.groupby(['Primary Type','Arrest']).size().reset_index(name='counts')
```

```python
df4=df3.pivot(index='Primary Type',columns='Arrest',values='counts')
```

```python
df4.to_csv('Chicago_Crimes_2012_to_2016_light_arrest_prob.csv', index=True)
```

```python
df4
```

```python
df=crime_light.groupby(['month','year']).size().reset_index(name='counts')
```

```python
df2=crime_light.groupby(['year','Primary Type']).size().reset_index(name='counts')
```

```python
df2
```

```python
df2_2012=df2.loc[df2['year'] == 2012]
```

```python
df2_2012['probability']=df2_2012['counts']/df2_2012['counts'].sum()
```

```python
years=[2012,2013,2014,2015,2016]
frames=[]
for year in years:
    df_year=df2.loc[df2['year'] == year]
    df_year['probability']=df_year['counts']/df_year['counts'].sum()
    frames.append(df_year)
    df2_prob = pd.concat(frames)
```

```python
df2_prob
```

```python
df2_prob.to_csv('Chicago_Crimes_2012_to_2016_light_donut.csv', index=False)
```

```python
df=df.pivot(index='month',columns='year',values='counts')
```
